```python
import pandas as pd
import numpy as np

# Create a test dataframe with Employee Name, Department, and Years of Experience columns
data = pd.DataFrame({
    "Employee Name": ['Alice', 'Bob', 'Charlie', 'David'],
    "Department": ["HR", "Engineering", "Marketing", "Finance"],
    "Years of Experience": [5, 10, 3, 15]
})

# Display the dataframe
data
```

```python
# Add a new column 'Years to Retirement' using a lambda function
data['Years to Retirement'] = data['Years of Experience'].apply(lambda x: 35 - x)

# Display the updated dataframe
data
```

```python
# Create a flag column using a lambda function
data['is_in_HR_or_Finance'] = data['Department'].apply(
    lambda x: True if x in ['HR', 'Finance'] else False
)

# Display the updated dataframe
data
```
